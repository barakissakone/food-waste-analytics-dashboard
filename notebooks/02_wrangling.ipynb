```python
import pandas as pd
import numpy as np

df =pd.read_csv('../data/raw/FoodWasteAI.csv')
```

```python
#Compter les valeurs manquantes dans chaque colonne.
df.isnull().sum()
```

```python
#verification  et suppression des doublons 

duplicates = df.duplicated().sum()

print("Nombre de doublons :", duplicates)

if duplicates > 0:
    df = df.drop_duplicates()
    print(" Doublons supprimés")
else:
    print(" Aucun doublon trouvé")
```

```python
#Renommer les colonnes en minuscules
df.columns = df.columns.str.lower()
```

```python
#Afficher les informations sur le DataFrame
df.info()
```

```python
#Enregistrer le DataFrame nettoyé dans un nouveau fichier CSV
df.to_csv("../data/clean/FoodWasteAI_clean.csv", index=False)
```
