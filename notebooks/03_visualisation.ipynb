```python
#importation des bibliothèques
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#importer notre dataset
df=pd.read_csv("../data/clean/FoodWasteAI_clean.csv")
```

```python
df.info()
```

```python
#voir l'evolution de notre dataset
plt.hist(df["combined_figures_kg_capita_year"])
plt.show()
```

```python
#Comment les valeurs évoluent-elles au fil du temps ? 
plt.plot(df["household_estimate_kg_capita_year"])
plt.show()
```

```python
#Trouver les 10 pays qui gaspillent le plus de nourriture puis afficher ça dans un graphique.
Top10 = df.sort_values(by="combined_figures_kg_capita_year", ascending=False).head(10)
plt.bar(
    Top10["country"],
    Top10["combined_figures_kg_capita_year"],
)
plt.xticks(rotation=90)
plt.show()
```

```python
#Voir la répartition des régions dans le dataset
region_count=df["region"].value_counts()
plt.pie(region_count,
        labels=region_count.index,
        autopct="%1.1f%%",
        )

plt.show()
```

```python
#Voir la corrélation entre les différentes variables numériques du dataset

correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#voir la relation entre les déchets ménagers et les déchets de détail
plt.scatter(
    df["household_estimate_tonnes_year"],
    df["retail_estimate_tonnes_year"]
)

plt.xlabel("Household (tonnes/an)")
plt.ylabel("Retail (tonnes/an)")
plt.title("Household vs Retail")

plt.show()
```
